```python
#Chargement du fichier csv
df_prod = pd.read_csv("billets_production.csv", sep=None, engine='python', header=0)
```

#Etape 2 - Sélection des données

```python
#Vérification des colonnes disponibles
df_prod.head()
```

```python
colonnes_modele = ["height_left", "height_right", "margin_low", "margin_up", "length"]

# Vérification que toutes les colonnes sont présentes
manquantes = [col for col in colonnes_modele if col not in df_prod.columns]
if manquantes:
    print("⚠️ Colonnes manquantes :", manquantes)
else:
    print("✓ Toutes les colonnes nécessaires sont présentes.")

# Construction de X_prod avec les colonnes dans le bon ordre
X_prod = df_prod[colonnes_modele]
```

```python
#Sélection des mêmes variables explicatives que celles utilisées pour l'entraînement
X_prod = df_prod[['height_left', 'height_right', 'margin_low', 'margin_up', 'length']]
```

#Etape 3 - Chargement du pipeline

```python
#Chargement du pipeline sauvegardé
pipeline_loaded = joblib.load("meilleur_modele.pkl")
```

#Etape 4 - Prédiction

```python
#Prédictions
y_pred_prod = pipeline_loaded.predict(X_prod)
y_proba_prod = pipeline_loaded.predict_proba(X_prod)[:,1]
```

```python
#Ajout des résultats au DataFrame
df_prod['prediction'] = y_pred_prod
df_prod['proba'] = y_proba_prod
```
